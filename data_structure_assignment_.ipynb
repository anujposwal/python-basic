Q8 What is a hash table, and how does it relate to dictionaries in Python?
* A hash table is a data structure that stores data in key–value pairs. It uses a hash function to convert a key into an index (address) in memory, so data can be found very fast (almost O(1) time).
* Relation to Python Dictionary :
  A Python dictionary (dict) is implemented using a hash table.
   * Keys in a dictionary are hashed
   * The hash value decides where the data is stored
   * This makes dictionary operations like search, insert, and delete very fast

Q10. Explain why strings are immutable in Python.


* Strings in Python are immutable, which means their contents cannot be changed after they are created. If you modify a string, Python creates a new string instead of changing the original one. This improves memory efficiency, enhances security, and makes strings reliable for use as dictionary keys and set elements.

Q11. What advantages do dictionaries offer over lists for certain tasks?


* Dictionaries offer several advantages over lists when data needs to be accessed using unique keys instead of index positions.

Advantages:

* Faster data retrieval using keys.
* Stores data in key-value pairs, making it easier to organize related information.
* Allows quick updating and deletion of values.
* Suitable for storing structured data such as student records or employee details.

Q14. How does the 'in' keyword work differently for lists and dictionaries?



* The in keyword is used to check whether an element exists in a collection.

* In a list, it checks whether a value exists.

Q15. Can you modify the elements of a tuple? Explain why or why not.



* No, the elements of a tuple cannot be modified because tuples are immutable. Once a tuple is created, its values cannot be changed, added, or removed. If changes are needed, a new tuple must be created.

Q16. What is a nested dictionary, and give an example of its use case?

Q17. Describe the time complexity of accessing elements in a dictionary.



* The average time complexity for accessing an element in a Python dictionary is O(1) (constant time). This is because dictionaries use a hash table, which allows fast lookup using keys.

* In rare cases, due to hash collisions, the time complexity may become O(n), but under normal conditions, dictionary access is very efficient.

Q18. In what situations are lists preferred over dictionaries?



Lists are preferred when:

* The order of elements is important.
* Duplicate values need to be stored.
* Data is accessed using index positions.
* The collection contains similar types of items

Q19. Why are dictionaries considered unordered, and how does that affect data retrieval?



* Dictionaries store data as key-value pairs rather than by index. Traditionally, they were considered unordered because elements were not arranged based on insertion order. In Python 3.7 and later, dictionaries preserve insertion order, but data is still accessed using keys, not index positions.

* Since dictionaries use keys for retrieval, searching for a value is fast and efficient.
