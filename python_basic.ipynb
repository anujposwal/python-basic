Q 6 Why are lists mutable, but tuples are immutable ?
 * Lists and tuples are both data structures in Python, but they have different use cases and design principles.

Lists are mutable because they're designed for dynamic modification, such as:

1. Adding or removing elements
2. Changing element values
3. Inserting or deleting elements at specific positions

This mutability makes lists suitable for applications where data needs to be frequently updated.

Tuples, on the other hand, are immutable because they're designed for:

1. Representing fixed, constant data
2. Providing a lightweight, efficient data structure
3. Ensuring data integrity and thread safety

Q 8  What are logical operators in Python ?

1. and (Logical AND)

- Returns True if both conditions are true.
- Returns False if either condition is false.

Example: x = 5 , y = 10, x > 0 and y > 5 ; returns True

2. or (Logical OR)

- Returns True if either condition is true.

Example: x = 5 , y = 0 , x > 0 or y > 5 ; returns True

Q 10 What is the difference between implicit and explicit type casting ?
 * In Python, type casting is the process of converting one data type into another. There are two types of type casting:

1. Implicit Type Casting: Python automatically converts one data type to another when necessary.

Example: a = 5; b = 2.5; c = a + b (Python converts a to float)

2. Explicit Type Casting: You manually convert one data type to another using built-in functions like:
    - int(): Converts to integer
    - float(): Converts to floating-point number
    - str(): Converts to string
    - bool(): Converts to boolean

Examples:
* x = int(3.7) returns 3
* y = float(5) returns 5.0
* z = str(123) returns "123"

Type casting helps ensure compatibility and accuracy when working with different data types in Python.

Q 12  How does the elif statement work ?

 we  can use elif statement like this ;

1. The if statement is evaluated first.
2. If the if condition is true, the code block under it is executed.
